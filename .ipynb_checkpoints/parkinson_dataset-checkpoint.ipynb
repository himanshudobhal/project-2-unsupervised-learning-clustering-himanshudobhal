```python
from sklearn.preprocessing import StandardScaler
import numpy as np
X = pk_dataset.drop('status',axis=1)
X = X.drop('name',axis=1)

sc = StandardScaler()
X_std =  sc.fit_transform(X)          
cov_matrix = np.cov(X_std.T)
```

```python
cov_matrix
```

```python
eig_vals, eig_vecs = np.linalg.eig(cov_matrix)
```

```python
total = eig_vals.sum()
variance_explained = [(each / total)*100 for each in eig_vals]
```

```python
variance_explained.sort(reverse=True)
```

```python
np.cumsum(variance_explained)
```

```python
plt.plot(np.cumsum(variance_explained))
```

```python
from sklearn.decomposition import PCA

pca_model = PCA(n_components=22)
pca_model.fit(X_std)
```

```python
pca_model.components_ 
```

```python
df_comp = pd.DataFrame(pca_model.components_,columns=X.columns.values)
```

```python
plt.figure(figsize=(20,6))
sns.heatmap(df_comp,cmap='plasma', annot=True)
```

```python
X_dim = pca_model.transform(X_std)
```
