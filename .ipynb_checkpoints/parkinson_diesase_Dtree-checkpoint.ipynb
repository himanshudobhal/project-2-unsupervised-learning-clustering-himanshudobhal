### PCA for the data set

```python
from sklearn.preprocessing import StandardScaler
```

```python
pdis_1.shape
```

```python
X = pdis_1.drop('status',axis=1)
```

```python
X = X.drop('name',axis=1)
```

```python
X.shape
```

```python
import numpy as np
```

```python
sc = StandardScaler()
X_std =  sc.fit_transform(X)          
cov_matrix = np.cov(X_std.T)
print('Covariance Matrix \n%s', cov_matrix)
```

```python
#Compute the eigen vectors and values using linear algebra lambda decomposition

eig_vals, eig_vecs = np.linalg.eig(cov_matrix)
print('Eigen Vectors \n%s', eig_vecs)
print('\n Eigen Values \n%s', eig_vals)
```

### Calculate the CUM Explained Variance

```python
total = eig_vals.sum()
variance_explained = [(each / total)*100 for each in eig_vals]
```

```python
variance_explained.sort(reverse=True)
```

```python
type(variance_explained)
```

```python
np.cumsum(variance_explained)
```

```python
plt.plot(np.cumsum(variance_explained))
```

### 1st 8 variables contributes upto 95% of variation

```python
from sklearn.decomposition import PCA

pcamodel = PCA(n_components=22)
pcamodel.fit(X_std)
```

```python
pcamodel.components_ 
```

```python
df_comp = pd.DataFrame(pcamodel.components_,columns=X.columns.values)
```

```python
plt.figure(figsize=(20,6))
sns.heatmap(df_comp,cmap='plasma', annot=True)
```
